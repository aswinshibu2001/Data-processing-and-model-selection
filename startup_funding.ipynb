### finding the junk value in date column 

```python
import re
```

```python
pattern = r'\d\d/\d\d/\d\d\d\d'
```

```python
type(df['Date dd/mm/yyyy'][0])
```

```python
junk_index=[]
for index, row in enumerate(df['Date dd/mm/yyyy']):
    if re.search(pattern,row) == None:
        print(index,row)
        junk_index.append(index)
```

```python
df.loc[junk_index]
```

```python
df.drop(junk_index,inplace=True)
```

```python
df.loc[junk_index]
```
